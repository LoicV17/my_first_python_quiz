```python
import time

questions_dic = {
    "Quelle est la capitale de l'Italie ?": "rome",
    "Qui est le nouveau président des USA?": "donald trump",
    "Quel est le prénom de Messi?": "lionel"
}


def count_error(n):
    if n == 1 :
        # Si l'utilisateur a fait 1 erreur, affiche qu'il lui reste 2 chances
        print("You have 2 chances left")
    elif n == 2 :
        # Si l'utilisateur a fait 2 erreurs, affiche qu'il lui reste 1 chance
        print("You have 1 chance left")
    elif n == 3 :
        # Si l'utilisateur a fait 3 erreurs, affiche qu'il a perdu et arrête le quiz
        print("Oh no! you lost the quiz")
        raise ValueError("Fin du quiz")  # Lève une erreur pour arrêter le quiz*

def ask_questions(question, correct_answer, nb_error) :
    answer = input(question)
    while answer.lower() != correct_answer :
        print("Your answer :", answer)
        print("Too bad ! C'est une mauvaise réponse")
        nb_error += 1
        count_error(nb_error)
        time.sleep(2)
        answer = input(question)

    print("Your answer :", answer)
    print("Good job ! C'est une bonne réponse")
    print("")

# initialisation du compteur d'erreur
nb_error = 0

# initialisation du n° de la question
i = 1

# Affichage du début du quiz
print("Here you have the quiz")
print("________________")
print("")

# Parcours des questions et vérification des réponses
for question, correct_answer in questions_dic.items():
        print("Question:",i)
        time.sleep(2)
        ask_questions(question, correct_answer, nb_error) # Réassigner les valeurs nb_error et i en dehors de la boucle
        i += 1

# Message de fin du quiz
print("________________")
print("Well Done! - You are the quiz winner!!")

    


```
